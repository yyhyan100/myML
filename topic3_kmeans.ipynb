```python
# generate 3D points (data with 3 features)
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt 
X, y = make_blobs(n_samples=100, centers=3, n_features=3,cluster_std=0.5,random_state=0,)
#print(X)
#print(y)
fig=plt.Figure()
ax = plt.axes(projection='3d')
ax.scatter3D(X[y==0][:,0],X[y==0][:,1],color="r")
ax.scatter3D(X[y==1][:,0],X[y==1][:,1],color="b")
ax.scatter3D(X[y==2][:,0],X[y==2][:,1],color="g")
```

```python
from myML.clustering import kmeans 
kms=kmeans()
y_predict=kms.fit(X,k=3)
fig=plt.Figure()
ax = plt.axes(projection='3d')
ax.scatter(X[y_predict==0][:,0],X[y_predict==0][:,1],color="r")
ax.scatter(X[y_predict==1][:,0],X[y_predict==1][:,1],color="b")
ax.scatter(X[y_predict==2][:,0],X[y_predict==2][:,1],color="g")
```

```python
print(kms.get_centroid())
```

```python
# now, let's try the Kmeans++ algorithm
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt 
X, y = make_blobs(n_samples=100, centers=3, n_features=3,cluster_std=0.5,random_state=0,)
#print(X)
#print(y)
fig=plt.Figure()
ax = plt.axes(projection='3d')
ax.scatter(X[y==0][:,0],X[y==0][:,1],color="r")
ax.scatter(X[y==1][:,0],X[y==1][:,1],color="b")
ax.scatter(X[y==2][:,0],X[y==2][:,1],color="g")
```

```python
from myML.clustering import kmeanspp
kmspp=kmeanspp()
y_predict=kmspp.fit(X,k=3)
fig=plt.Figure()
ax = plt.axes(projection='3d')
ax.scatter(X[y_predict==0][:,0],X[y_predict==0][:,1],color="r")
ax.scatter(X[y_predict==1][:,0],X[y_predict==1][:,1],color="b")
ax.scatter(X[y_predict==2][:,0],X[y_predict==2][:,1],color="g")
```

```python
# plot the init centers
fig=plt.Figure()
ax = plt.axes(projection='3d')
print(kmspp.get_centroid())
ax.scatter(kmspp.get_centroid()[0],kmspp.get_centroid()[1])
```
